```python
def calculate_entropy(df, target_column_index):
    target_column = df[:, target_column_index]
    _, counts = np.unique(target_column, return_counts = True)
    probabilities = counts/counts.sum()
    
    entropy = sum(probabilities*-np.log2(probabilities))
    
    return entropy
```

```python
def calc_information_gain(target_column, databelow, dataabove):
    
    information_gain = calculate_entropy(target_column) - calculate_entropy(databelow) - calculate_entropy(dataabove)
    
    return information_gain
```
